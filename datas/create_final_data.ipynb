```python
import pandas as pd
import numpy as np
```

```python
df_culture = pd.read_csv('culture.csv')
df_sport = pd.read_csv('sport.csv')
df_legal = pd.read_csv('legal.csv')
df_polictics = pd.read_csv('politics_data.csv')
```

```python
df_culture['label'] = 'Culture'
df_culture = df_culture.rename(columns={df_culture.columns[0]: 'text'})

df_sport['label'] = 'Sport'
df_sport = df_sport.rename(columns={df_sport.columns[0]: 'text'})

df_legal['label'] = 'Legal'
df_legal = df_legal.rename(columns={df_legal.columns[0]: 'text'})

df_polictics['label'] = 'Politics'
df_polictics = df_polictics.rename(columns={df_polictics.columns[0]: 'text'})
```

```python
final_data = pd.concat([df_culture, df_sport, df_polictics], ignore_index=True)
```

```python
final_data = final_data[['label', 'text']]
```

```python
final_data.to_csv('final_data.csv', index=False, encoding='utf-8-sig')
```
